```python
import pandas as pd
```

```python
ev_chargers_by_city_df = pd.read_csv("ev_cities_clean.csv")

ev_chargers_by_city_df['City'] = ev_chargers_by_city_df['City'].str.lower().str.strip()

result_ev_chargers_by_city_df = ev_chargers_by_city_df.groupby('City', as_index=False).agg({
    'Fuel Type Code': 'first',  
    'State': 'first',  
    'EV Total Port Count (Beta)': 'sum'
})

result_ev_chargers_by_city_df.to_csv("ev_chargers_by_city.csv", index=False)

```

```python
evchargers_bycity_final_df = pd.read_csv("ev_chargers_by_city.csv")

evchargers_bycity_final_df.head(200)
```

```python
evchargers_bycity_final_df['City'] = evchargers_bycity_final_df['City'].str.capitalize()

evchargers_bycity_final_df.head(200)

```

```python
file1_df = pd.read_csv("Median_Income_by_City.csv")
file2_df = pd.read_csv("Census_Population_By_City.csv")

file1_df['City'] = file1_df['City'].str.capitalize()
file2_df['City'] = file2_df['City'].str.capitalize()
evchargers_bycity_final_df['City'] = evchargers_bycity_final_df['City'].str.capitalize()

merged_df = pd.merge(evchargers_bycity_final_df, file1_df, on='City', how='outer', suffixes=('_evchargers', '_file1'))
merged_df = pd.merge(merged_df, file2_df, on='City', how='outer', suffixes=('_evchargers', '_file2'))


merged_df.head(200)
```

```python
merged_df.to_csv("combined.csv", index=False)
```

```python
columns_to_keep = ['City', 'State_file1', 'EV Total Port Count (Beta)', 'Median Income (16 and over)', 'Total Population']
merged_df = merged_df.dropna(subset=columns_to_keep)

merged_df.head(200)
```

```python
merged_df.to_csv("combined2.csv", index=False)
```

```python
merged_df.columns
```

```python
merged_df.drop (columns=[ 'Fuel Type Code', 'State_evchargers', 'State_file1']
    
, inplace=True)

merged_df.head (200)
```

```python
column_order = ['City', 'State', 'EV Total Port Count (Beta)', 'Median Income (16 and over)', 'Total Population']

merged_df = merged_df[column_order]

merged_df.head (200)
```

```python
merged_df.columns
```

```python
new_column_names = ['City', 'State', 'Total EV Chargers', 'Median Income (16 and over)', 'Total Population']


merged_df.columns = new_column_names

merged_df.head (200)
```

```python
# Convert columns to integers
columns_to_convert = ['Total EV Chargers', 'Median Income (16 and over)', 'Total Population']
merged_df[columns_to_convert] = merged_df[columns_to_convert].astype(int)

merged_df.head (200)
```

```python
# Save the merged DataFrame to a new CSV file
merged_df.to_csv("combined_final.csv", index=False)
```
